```r
#KEYWORD - #flu2018

library(sp)
library(maps)
library(maptools)
library(twitteR)
library(plyr) 
library(ggplot2)
library(ggmap)
library(data.table)

stateconv <- function(curateddata) {
    states <- map('state', fill=TRUE, col="transparent", plot=FALSE)
    IDs <- sapply(strsplit(states$names, ":"), function(x) x[1])
    statescurated <- map2SpatialPolygons(states, IDs=IDs, proj4string=CRS("+proj=longlat +datum=WGS84"))
    pointscurated <- SpatialPoints(curateddata, proj4string=CRS("+proj=longlat +datum=WGS84"))
    indices <- over(pointscurated, statescurated)
    stateNames <- sapply(statescurated@polygons, function(x) x@ID)
    stateNames[indices]
    write.csv(stateNames[indices], file = "States-flu2018.csv")
}

tweets <- searchTwitter('#flu2018',n=10000)
tweetsdata = ldply(tweets, function(t) t$toDataFrame())

screennames <- tweetsdata$screenName
screennamesdata <- twListToDF(lookupUsers(screennames))

location <- screennamesdata[,12]
location <- iconv(location, "latin1", "ASCII", sub="")
location[location==""] <- NA
location <- na.omit(location)
locationdata <- data.frame(location)
    
curated <- geocode(location)
curated <- na.omit(curated)
curateddata <- data.frame(curated)
    
write.csv(curated, file = "Tweets-flu2018.csv")#Choose States-flu2018.csv

stateconv(curateddata)

statesfile <- read.csv(file.choose())
occurences<-table(unlist(statesfile$x))
write.csv(occurences, file = "Occurences-flu2018.csv")

StateDataforMap <- data.table(read.csv(file.choose()))#Choose Occurences-flu2018.csv
StateDataforMap$region <- tolower(StateDataforMap$Var1)
latlongdata <-data.table(map_data("state"))
setkey(latlongdata,region)
setkey(StateDataforMap,region)
StateLoc<-latlongdata[StateDataforMap]
map <- ggplot() + geom_polygon(data = StateLoc,aes(x = long,y = lat, group = group,fill=Freq))+ labs(fill = "Activity Level" 
                            ,title = "Twitter HeatMap", x="", y="")+coord_fixed(1.3)
finalmap <- map + theme_bw() 
finalmap + scale_y_continuous(breaks=c()) + scale_x_continuous(breaks=c()) + theme(panel.border =  element_blank())
finalmap + scale_fill_continuous(low = "green", high = "red") 
```
